```
import math

def hitung_lingkaran(jari_jari):

    luas = math.pi * (jari_jari ** 2)
    keliling = 2 * math.pi * jari_jari
    return luas, keliling


# Contoh penggunaan fungsi
jari_jari = 7
luas_lingkaran, keliling_lingkaran = hitung_lingkaran(jari_jari)

print(f"Jari-jari lingkaran: {jari_jari}")
print(f"Luas lingkaran: {luas_lingkaran}")
print(f"Keliling lingkaran: {keliling_lingkaran}")
```
